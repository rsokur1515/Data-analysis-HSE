```python
import numpy as np
from scipy import stats as sts
import random
np.random.seed(1000) #задаю зерно генератора для всех рандомных генераций
                                                                          #WARNING!#
#Я сам не до конца разобрался как работает эта штука. В теории, все генерации должны быть одинаковыми, но по факту все работает нормально. Если вдруг
# что-то будет работать очень страннно, попробуй удалить эту штуку. Спасибо!
```

```python
t = np.random.choice([5, 10], 144, p=[0.5, 0.5]) #генерируем рандомное время автобус (генерируем на всякий случай 720, чтобы вариантов времени хватило, даже
# если бы каждый интервал между автобусами был бы равен 5-и минутам)
t
```

```python
np.cumsum(t[:5]) #тестирую cumsum, чтобы в последующем определять, какой автобус будет конечным (чтобы укладываться по времени работы автобусов)
```

```python
def bus_per_day(): #задаю функцию, которая определяет, сколько автобусов будет отходить от станции за день (один эксперимент)
  t = np.random.choice([5, 10], 144, p=[0.5, 0.5])
  t = np.cumsum(t)
  return len(t[t<=720])
bus_per_day()
```

```python
res_exp_a = [] #провожу 10**4 подобных экспериментов
for k in range(10**4):
  res_exp_a.append(bus_per_day())
res_exp_a[:10] # результаты сохраняю в данную переменную
```

### Задача 3 (Парадокс Берксона)

Предположим, что результаты ЕГЭ школьников по русскому и математике независимы и хорошо аппроксимируются 
нормальным распределением с ожиданием 60 и стандартным отклонением 10. 

УШЭ (Урюпинская Школа Экономики) ныне престижна и забирает себе всех школьников набравших более $n$ баллов в сумме по русскому и математике. 

Для каждого $n$ от $80$ до $160$ с шагом $5$ случайно создайте $10^4$ школьников и поделите их на прошедших и не прошедших в УШЭ.


__а) [5 баллов]__ Постройте график числа прошедших в УШЭ в зависимости от $n$. 

__б) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди прошедших в УШЭ.

__в) [5 баллов]__ Постройте график выборочной корреляции между результатами по русскому и математике в зависимости от $n$ среди не прошедших в УШЭ.

a)

```python
norm_rv = sts.norm(60, 10) #задаем генератор нормального распределения
russian = norm_rv.rvs((17, 10**4)) #генерируем выборки с результатами по русскому языку
russian[np.where(russian < 0)] = 0 #все отрицательные баллы обращаем в 0, так как невозможно получить отрицательные баллы за ЕГЭ
russian = russian.astype('int32') # преобразую результаты за ЕГЭ по русскому в целые числа, так как получить дробный балл за ЕГЭ также невозможно
maths = norm_rv.rvs((17, 10**4)) #все то же самое, но для математики
maths[np.where(maths < 0)] = 0
maths = maths.astype('int32') # преобразую результаты за ЕГЭ по математике в целые числа, так как получить дробный балл за ЕГЭ также невозможно
results = russian + maths #суммарные баллы за русский и математику
results[0][:20]
```

```python
ball, res_exp_a = 80, [] #считаю количество поступивших для различного проходного балла
for k in range(17):
  res_exp_a.append(len(results[k][results[k]>=ball]))
  ball += 5
res_exp_a[:20]
```

```python
x = np.arange(80, 161, 5)
plt.figure(figsize=(9,6))
plt.title('Количество прошедших в ВУЗ в зависимости от проходного балла')
plt.xlabel('Требуемые баллы по ЕГЭ')
plt.ylabel('Количество поступивших')
plt.plot(x, res_exp_a, color = 'blue');
```

На данном графике можно увидеть зависимость количества поступивших в УШЭ абитуриентов в зависимости от проходных баллов по ЕГЭ. Логика взимозависимости следующая: чем больше требуемый балл, тем меньше поступивших в ВУЗ. Тем не менее, взаимосвзяь не линейная, так как график не похож на прямую. Исходя из иллюстрации, можно сделать вывод, что в целом зависимость практически прямая, однак0, чем ближе балл к краевым значениям: 80 и 160-и баллам, тем меньшее влияние оказывает увеличение или уменьшение баллов. Таким образом в районе 80-и баллов количество студентов уменьшается менее, чем на 500 человек при поднятии проходного на 10 пунктов. Однако в окружности 105-и баллов, аналогичное повышение баллов приводит к потере около 2-ух тысяч абитуриентов.

б)

```python
ball, res_exp_b = 80, [] #считаю корелляцию между результатами по русскому и математике для поступивших студентов
for k in range(17):
  res = maths[k] + russian[k]
  index = np.where(res>=ball)
  ball += 5
  res_exp_b.append(np.corrcoef(maths[k][index], russian[k][index])[0][1])
res_exp_b[16]
```

```python
x = np.arange(80, 161, 5)
plt.figure(figsize=(9,6))
plt.title('Выборочная корреляция в зависимости от проходного балла для поступивших в УШЭ')
plt.xlabel('Требуемые баллы по ЕГЭ')
plt.ylabel('Корреляция')
plt.plot(x, res_exp_b, color = 'red');
```

Исходя из заданных условий, согласно которым результаты студентов по русскому языку и метематике - независимые одинаково распределенные величины, корреляция между ними, должна быть около нуля с поправкой на ошибку выборки, однако исходя из графика мы видим, что по мере увеличения требуемых для поступления в УШЭ баллов, корреляция отклоняется от логически определенного показателя корреляции в отрицательную сторону. Таким образом, когда влияние дополнительного условия, которое заключается в поступлении в ВУЗ, имеет больший характер, корреляция между величинами уменьшается, начинаясь в 0-е, когда влияние данного фактора минимальное, так как, исходя из графика в пункте а) при низких значениях проходного балла, практически все студенты поступают в УШЭ, заканчивая меньше чем на -0,8, где влияние фактора максимальное (160 требуемых баллов). Из чего можно сделать вывод о том, что имеет место парадокс Берксона.

в)

```python
ball, res_exp_v = 80, [] #считаю корелляцию между результатами по русскому и математике для поступивших студентов
for k in range(17):
  res = maths[k] + russian[k]
  index = np.where(res<ball)
  ball += 5
  res_exp_v.append(np.corrcoef(maths[k][index], russian[k][index])[0][1])
res_exp_b[16]  
```

```python
x = np.arange(80, 161, 5)
plt.figure(figsize=(9,6))
plt.title('Выборочная корреляция в зависимости от проходного балла для НЕпоступивших в УШЭ')
plt.xlabel('Требуемые баллы по ЕГЭ')
plt.ylabel('Корреляция')
plt.plot(x, res_exp_v, color = 'green');
```

В пункте в) прослеживается аналогичная пункту б) логика, однако влияние условного фактора в виде поступления в УШЭ в данном случае играет тем меньшую роль, чем больше необходимый проходной балл, так как количество рассматриваемых результатов и выборок для оценки корреляции увеличивается, в связи с тем, что, опираясь на график из пункта а), чем больше проходной балл, тем больше НЕпоступивших людей. Так, чем меньше проходной балл, тем меньше НЕпоступивших абитуриентов, тем большее влияние оказывает условный фактор, и тем больше выборка отличается от генеральной совокупности, в результате чего при проходном балле в 80, корреляция достигает -0,8, а при 160- нулевая корреляция, какой она и должна быть.
